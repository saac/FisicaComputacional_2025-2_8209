$ x = y(a+x^2)$

$ y = \dfrac{b}{a+x^2}$

$\;$

```python
def f1(x,y,a):
    return y*(a+x**2)

def g1(x,b):
    return b/(a+x**2)
```

```python
a = 1
b = 2

x0,y0 = 3,4

x1 = f1(x0,y0,a)
y1 = g1(x0,b)

print(x1,",",y1)

x2 = f1(x1,1,a)
y2 = g1(x1,b)

print(x2,",",y2)

x3 = f1(x2,y2,a)
y3 = g1(x2,b)

print(x3,",",y3)
```
